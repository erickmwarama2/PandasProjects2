```python
import pandas as pd
import numpy as np
```

```python
pop = pd.read_csv('state-population.csv')
areas = pd.read_csv('state-areas.csv')
abbrevs = pd.read_csv('state-abbrevs.csv')
```

```python
pop.head()
```

```python
areas.head()
```

```python
abbrevs.head()
```

```python
state_abbrev = pd.merge(areas, abbrevs, on='state')
state_abbrev.head()
```

```python
pop.head()
```

```python
state_abbrev_pop = pd.merge(state_abbrev, pop, left_on='abbreviation', right_on='state/region')
state_abbrev_pop.head()
```

```python
state_abbrev_pop['density'] = state_abbrev_pop['population'] / state_abbrev_pop['area (sq. mi)']
state_abbrev_pop.head()
```

```python
state_abbrev_pop_2010 = state_abbrev_pop[(state_abbrev_pop.year == 2010) & (state_abbrev_pop.ages == 'total')]
state_abbrev_pop_2010.head()
```
